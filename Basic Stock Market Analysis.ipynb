# Importing Libraries and Data

```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as npd
import pandas as pd
%matplotlib inline
```

```python
start = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2022, 1, 1)
```

```python
tcs = web.DataReader("TCS.NS", "yahoo", start, end)
infosys = web.DataReader("INFY.NS", "yahoo", start, end)
wipro = web.DataReader("WIPRO.NS", "yahoo", start, end)
hcltech = web.DataReader("HCLTECH.NS", "yahoo", start, end)
```

```python
tcs.to_csv("TCS_Stock.csv")
infosys.to_csv("Infosys_Stock.csv")
wipro.to_csv("Wipro_Stock.csv")
hcltech.to_csv("HCLTech_Stock.csv")
```

# Daily Percentage Change

```python
tcs['returns'] = (tcs['Close']/tcs['Close'].shift(1)) - 1
infosys['returns'] = (infosys['Close']/infosys['Close'].shift(1)) - 1
hcltech['returns'] = (hcltech['Close']/hcltech['Close'].shift(1)) - 1
wipro['returns'] = (wipro['Close']/wipro['Close'].shift(1)) - 1
```

### Volatility

#### Plotting Histogram of Daily Percantage change

```python
f, ax = plt.subplots(2, 2, figsize=(14,8), sharex=True)
ax[0,0].hist(tcs['returns'], bins = 100, label = 'TCS', alpha = 0.5, color = 'r')
ax[0,0].set_title("TCS")

ax[0,1].hist(infosys['returns'], bins = 100, label = 'INFY', alpha = 0.5, color = 'g')
ax[0,1].set_title("INFY")

ax[1,0].hist(hcltech['returns'], bins = 100, label = 'HCLTECH', alpha = 0.5, color = 'b')
ax[1,0].set_title("HCLTECH")

ax[1,1].hist(wipro['returns'], bins = 100, label = 'WIPRO', alpha = 0.5, color = 'y')
ax[1,1].set_title("WIPRO")

f.show()
```

#### Comparing frequency distribution of daily percentage change

```python
tcs['returns'].plot(kind = 'kde', label = 'TCS', figsize = (10, 6), color='r')
infosys['returns'].plot(kind = 'kde', label = 'INFY')
hcltech['returns'].plot(kind = 'kde', label = 'HCLTECH')
wipro['returns'].plot(kind = 'kde', label = 'WIPRO')
plt.legend()
plt.show()
```

#### Comparing them in a box plot

```python
box_df = pd.concat([tcs['returns'], infosys['returns'], hcltech['returns'], wipro['returns']], axis = 1)
box_df.columns = ['TCS Returns', 'INFY Retunrs', 'HCLTECH Returns', 'WIRPO Returns']

box_df.plot(kind = 'box', figsize = (16,6))
```

# Cummulative Return

```python
tcs['Cummulative Return'] = (1 + tcs['returns']).cumprod()
infosys['Cummulative Return'] = (1 + infosys['returns']).cumprod()
hcltech['Cummulative Return'] = (1 + hcltech['returns']).cumprod()
wipro['Cummulative Return'] = (1 + wipro['returns']).cumprod()
```

```python
tcs['Cummulative Return'].plot(label = 'TCS', figsize = (15, 7))
infosys['Cummulative Return'].plot(label = 'INFY')
hcltech['Cummulative Return'].plot(label = 'HCLTECH')
wipro['Cummulative Return'].plot(label = 'WIPRO')
plt.legend()
plt.show()
```

# RSI

```python
delta = tcs['Close'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
ema_up = up.ewm(com=13, adjust=False).mean()
ema_down = down.ewm(com=13, adjust=False).mean()
rs = ema_up/ema_down
tcs['RSI'] = 100 - (100/(1 + rs))
# Skip first 14 days to have real values
#tcs = tcs.iloc[14:]
#print(tcs)
fig, (ax1, ax2) = plt.subplots(2, figsize=(14,8))
ax1.get_xaxis().set_visible(False)
fig.suptitle('TCS')
tcs['Close'].plot(ax=ax1)
ax1.set_ylabel('Price')
tcs['RSI'].plot(ax=ax2)
ax2.set_ylim(0,100)
ax2.axhline(30, color='r', linestyle='--')
ax2.axhline(70, color='r', linestyle='--')
ax2.set_ylabel('RSI')
plt.show()
```
